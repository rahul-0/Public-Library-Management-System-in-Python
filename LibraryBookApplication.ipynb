```python
class Author:
	'''
		The __init()__ method of the Authors accepts parameters for all instance variables.
	'''
	def __init__(self, aunm, bkgn):
		self.__author_name = aunm
		self.__book_genre = bkgn

	#Define the accessor and mutator methods
	def get_name(self):
		return self.__author_name
		
	def get_genre(self):
		return self.__book_genre
		
	def set_name(self, aunm):
		self.__author_name = aunm
		
	def set_genre(self, bkgn):
		self.__book_genre = bkgn
		
	'''
		Redefine the __str__() method and return a string that contains the values
		of the two instance variables of the Author object
	'''
	def __str__(self):
		return "Author Name: " + self.__author_name + "; Book Genre: " + self.__book_genre 
```

```python
'''
The parent class must be imported
'''
from librarybook import LibraryBook
class TextBook(LibraryBook): #The TextBook class inherits from the LibraryBook class

	'''
		The __init()__ method of the TextBook class accepts parameters for all instance variables 
		including the ones needed for the LibraryBook class as well as the ones for the
		Author class.   Note that the __init__() method defined below calls the superclass
		__init__() method passing to it all arguments required by the LibraryBook class as
		well as the Author class. 
	'''
	def __init__(self, bkid, bknm, bkprice,noCopies,bkauthor, sub, disc):
		super().__init__(bkid, bknm, bkprice,noCopies,bkauthor)
		self.__discount = disc
		self.__subject = sub
		
	# Write all the accessor and mutator methods for the instance variables specific to the TextBook class
	def get_subject(self):
		return self.__subject
		
	def set_subject(self, sub):
		self.__subject = sub

	def get_discount(self):
		return self.__discount
		
	def set_discount(self, disc):
		self.__discount = disc
	
	'''
		The compute_revenue() method defined below will override the compute_revenue() method 
		fdefined in the LibraryBook class for all objects which are instances of TextBook
	'''
	def compute_revenue(self):
		nc = self.get_num_copies()
		if nc > 15:
			return (self.__discount/100) * nc * super().get_book_price()
		else:
			return nc * super().get_book_price()
	
	'''
		The __str__() method first calls the super class's __str__() method and then concatenates
		to the string that is returned, the values of all instance variables that are specific to the TexBook class.
	'''
		
	def __str__(self):

		childData = super().__str__()
		childData = "\nDetails of this Textbook are:" + childData
		childData += "\nSubject: " + self.__subject + '; Discount: ' + '{0:0.2f}'.format(self.__discount)
		return childData
```

```python
'''
The parent class must be imported
'''
from librarybook import LibraryBook
class Fiction(LibraryBook):
	'''
		The init() method of the Fiction class accepts parameters for all instance variables 
		including the ones needed for the LibraryBook class as well as the ones for the
		Author class.  Note that the __init__() method defined below calls the superclass
		__init__() method passing to it all arguments required by the LibraryBook class as
		well as the Author class.
	'''
	def __init__(self, bkid, bknm, bkprice,noCopies,bkauthor, isHC):
		super().__init__(bkid, bknm, bkprice,noCopies,bkauthor)
		self.__isHardCover = isHC

	# Write all the accessor and mutator methods for the instance variables specific to the TextBook class
	def get_isHardCover(self):
		return self.__isHardCover
		
	def set_isHardCover(self, isHC):
		self.__isHardCover = isHC

	'''
		The compute_revenue() method defined below will override the compute_revenue() method 
		fdefined in the LibraryBook class for all objects which are instances of the Fiction class
	'''		
	def compute_revenue(self):
		nc = self.get_num_copies()
		if self.__isHardCover: 
			return 3 * nc * super().get_book_price()
		else:
			return nc * super().get_book_price()

	'''
		The __str__() method first calls the super class's __str__() method and then concatenates
		to the string that is returned, the values of all instance variables that are specific to the Fiction class.
	'''			
	def __str__(self):

		childData = super().__str__()
		childData = "\nDetails of this Fiction Book are:" + childData
		childData += "\nIs the Book Hard Cover?: " + self.__isHardCover
		return childData
```

```python
'''
	- First import all the classes that will be used by this application.
	- In addition, import the pickle module since you will be writing and reading objects of
	a class to and from a file.
	- Also import the sys module to enable graceful termination of your program.
'''	
from librarybook import LibraryBook
from fiction import Fiction
from author import Author
from textbook import TextBook
import pickle
import sys

'''
Define the main() method.  This just calls the run_menu_options() method
'''
def main():
	run_menu_options()

'''
	The run_menu_options() first calls the read_file_data() method and stores the returned list into book_list.
	It then asks the user to enter one of three option, add a book, print the revenue or exit the application.
	Based on the option selected, the method calls run_option1, run_option2, or run_option3.
	Asking the user to select an option and executing the appropriate method is done repeatedly until the user
	exits the program.
'''	
def run_menu_options():
	book_list = read_file_data()
	sel = 1
	
	# Setup a while loop to get a response from the user and call the appropriate method.
	while sel >=  1 and sel <= 3:
		print("\n==== Menu ====")
		print("1. To add a Book")
		print("2. To print the names of all books together with their revenue")
		print("3. To exit program")
		
		# 	If the user enters an invalid value, a message is printed and the options are displayed again.
		try:
			sel = int(input("Your selection: "))
			if sel < 1 or sel > 3:
				raise ValueError
		except ValueError:
			print('You must enter an integer between 1 and 3!')
			print('Try again')
			sel = 1
			continue		

		if sel == 1:
			book_list.append(run_option1())
		elif sel == 2:
			if len(book_list) == 0: #This check displays an appropriate message if the user selects option 2 when there is no data in the file,
				print("No books in database!")
			else:
				run_option2(book_list) #call run_option2() by passing it a list containing the current book objects in the inventory
		elif sel == 3:
				run_option3(book_list) #call run_option3() by passing it a list containing the current book objects in the inventory

'''
	This method reads the binary file containing all the library books that are currently in the database
	and stores it in a list. Once all the book objects have been read into the list, it returns the list to the calling method.
'''			
def read_file_data():
	my_lst = [] #Create an empty list
	try:
		#Call the open command within a try block.  Note that the file is opend for binary read.
		fb = open('bookdata.dat', 'rb') 
	except FileNotFoundError:
		
		#If the file is not there (this will be the case, the first time the program is executed), the empty file is returned.
		return my_lst 
	
	#If there is no error reading the file, the rest of the code is executed.
	else: 
		try:
		
			#This is a repeated loop which will terminate only when the EOFError occurs.
			while (True): 
			
				#This statement reads the next object from the file and appends it to the list
				my_lst.append(pickle.load(fb))
		
		#When the EOFError occurs, the file is closed and the list (containing all the book objects that have just been read) is returned.
		except EOFError: 
			fb.close()
			return my_lst	
'''
	This method is executed each time the user selects option 1 (add a new book).
'''
def run_option1():
	sel = 0
	# Ask the user to enter the type of book to be added. 
	while sel < 1 or sel > 3: #Repeat the prompt until the user enters an valid value of 1, 2, or 3
		try:
			sel = int(input("Type of Library Book?(1-Text Book;2-Fiction) : "))
		except ValueError: 
			print('Wrong selection!')
			continue		
	# Call the get_book_input() method and store the values in the appropriate variables
	id, nm, pr, nc = get_book_input()
	
	# Call the get_author_input() method and store the values in the appropriate variables
	a_nm, g = get_author_input()
	an_auth = Author(a_nm, g) #Create an Author object
	
	#Call the get_textbook_input() or the get_fiction_input() method based on the user's selection
	#and then create the appropriate book object
	if sel == 1:
		sub, disc = get_textbook_input()
		a_book = TextBook(id, nm, pr, nc, an_auth,sub, disc)
	elif sel == 2:
		ishc = get_fiction_input()
		a_book = Fiction(id, nm, pr, nc, an_auth,ishc)
	return a_book #Return the book object to the calling method
'''
	This method is executed each time the user selects option 2 (print book name and revenue)
'''	
def run_option2(my_lst):
	
	#First print the column headins
	print("\nBook Name\t\t\tRevenue")
	print("=======================\t\t===========")
	
	'''
		Then, for each object in the list that was passed as an argument to this method, compute the
		revenue and print out the revenue together with the book name.
		Note the role of polymorphism here.  The list contains both TextBook objects as well as Fiction objects.
		When the compute_revenue() method is called, the appropriate method is executed based on whether the calling
		object is a Fiction object or a TextBook object.
	'''
	for obj in my_lst:
		rev = obj.compute_revenue()
		print('{0} \t\t${1:0.2f}'.format(obj.get_book_name(), rev))
'''
	This method is executed when the user selects option 3 (exit the program)
'''
def run_option3(my_lst):

	#The program first confirms that the user really wants to exit the program
	print("You chose to exit the program")
	yes_no = input("Are you sure (Y/N)? ")
	
	'''
		Then the bookdata.dat file is opened for output and all the book objects in the list are written out to the file.
		The file is then closed.
		Finally we exit the program.
	'''
	if yes_no.lower() == 'y':
		fb = open('bookdata.dat', 'wb')
		for obj in my_lst:
			pickle.dump(obj, fb)
		fb.close()
		sys.exit()	
'''
	Each of the four methods below accept input for the relevant intsnace variables.
	Exception handling is used to ensure that the user enters valid values for the variables.
'''	
def get_book_input():
	bk_id = input('Enter the book\'s id: ')
	bk_name = input('Enter the book\'s name: ')
	err_flg = True	
	while err_flg:
		try:
			bk_price = float(input('Enter the price of the book: '))
		except ValueError:
			print('\nEnter a numerical value for price!!')
		else:
			err_flg = False
	err_flg = True
	while err_flg:
		try:
			bk_no_copies = int(input('Enter the number of copies of this book: '))
		except ValueError:
			print('\nEnter an integer value for the number of copies!!')
		else:
			err_flg = False	
	return (bk_id, bk_name, bk_price, bk_no_copies)
	
def get_author_input():
	au_name = input('Enter the author\'s name: ')
	bk_gn = input('Enter the book genre the author specializes in: ')
	return (au_name, bk_gn)
	
def get_textbook_input():
	err_flg = True
	while err_flg:
		try:
			disc = float(input('Enter the quantity discount: '))
		except ValueError:
			print('\nEnter a numerical value for the discount!!')
		else:
			err_flg = False
	sub =input('Enter the subject the book deals with: ')
	return (sub, disc)
	
def get_fiction_input():
	err_flg = True
	while err_flg:
		try:ishc = bool(input('Does the book have a hard cover?: '))
		except ValueError:
			print('\nEnter a value of \'True\' or \'False\' only!!')
		else:
			err_flg = False
	err_flg = True

	return (ishc)
	


#Call the main() function to start the application
main()
```
